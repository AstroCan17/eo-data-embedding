into the documentation.

The notebook itself shows how to run a processor from the `eo_data_embedding` module.

## Import packages

Import all required packages in order to run the processor.

```python
from eo_data_embedding.computing.my_processor import MyProcessor
from eopf.product import EOProduct
from eopf.product.conveniences import init_product
```

## Init inputs

Create all required parameters as input for the processor.

```python
# Initialize inputs
product_a: EOProduct = init_product("MY_SUPER_PRODUCT.A")
product_a.attrs["id"] = "001"
product_b: EOProduct = init_product("ANOTHER_SUPER_PRODUCT.B")
product_b.attrs["id"] = "002"
```

## Run processor

Run the processor using the following code:

```python
# Run test
myProcessor = MyProcessor()
output = myProcessor.run(product_a, product_b, name="MY_PRODUCT.UNITTEST")
```
